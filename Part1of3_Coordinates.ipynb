```python
# Function for obtaining coordinates and returning latitude and longitude only
def getcoordinates(address):
    req = requests.get('https://developers.onemap.sg/commonapi/search?searchVal='+address+'&returnGeom=Y&getAddrDetails=Y&pageNum=1')
    resultsdict = eval(req.text)
    if len(resultsdict['results'])>0:
        return resultsdict['results'][0]['LATITUDE'], resultsdict['results'][0]['LONGITUDE']
    else:
        pass
```

```python
# for loop to keep track of progress during extraction, including the number of addresses that fail to return a coordinate
# append to dictionary for ease of conversion to dataframe later on

coordinatesdict= {'latitude': [], 'longitude': []}
count = 0
failed_count = 0
for address in shopping_malls:
    try:
        coordinates = getcoordinates(address)
        if coordinates is not None:
            count = count + 1
            print('Extracting',count,'out of',len(shopping_malls),'addresses')
            coordinatesdict['latitude'].append(coordinates[0])
            coordinatesdict['longitude'].append(coordinates[1])
        else:
            count = count + 1           
            failed_count = failed_count + 1
            print('Failed to extract',count,'out of',len(shopping_malls),'addresses')
            coordinatesdict['latitude'].append(None)
            coordinatesdict['longitude'].append(None)
    except:
        count = count + 1           
        failed_count = failed_count + 1
        print('Failed to extract',count,'out of',len(shopping_malls),'addresses')
        coordinatesdict['latitude'].append(None)
        coordinatesdict['longitude'].append(None)

print('Total Number of Addresses With No Coordinates',failed_count)
```

```python
# relating mall names to coordinates
coordinatesdict['malls'] = shopping_malls
coordinates_df=pd.DataFrame.from_dict(coordinatesdict)
coordinates_df.head()
```
